## 1) PMF plot

```python
from scipy.stats import randint
from scipy.stats import geom
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, ax = plt.subplots(1, 1)

p=0.2 
x = np.arange(geom.ppf(0.01, p), geom.ppf(0.90, p))
ax.plot(x, geom.pmf(x, p), 'bo', ms=8)
ax.vlines(x, 0, geom.pmf(x, p), colors='b', lw=5, alpha=0.5)
plt.xlabel('x')
plt.ylabel('P(x)')
plt.title('geom pmf')
```

## 2) CDF plot

```python
fig, ax = plt.subplots(1, 1)

low, high = 0, 10
x = np.arange(low, high, 0.01)
ax.plot(x, geom.cdf(x, p), 'b-', ms=8)
plt.xlabel('x')
plt.ylabel('P(x)')
plt.ylim([0,1.01])
plt.title('geom cdf')
```

## 3) 샘플 (sample) 발생

```python
r = geom.rvs(p, size=1000)
plt.hist(r, bins = np.arange(low+1,high+1)) 
plt.title("histogram") 
plt.show()
```
